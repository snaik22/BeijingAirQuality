```python
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()
#Plot rolling statistics:
    timeseries.plot(color='blue',label='Original')
    rolmean.plot(color='red', label='Rolling Mean')
    rolstd.plot(color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
test_stationarity(monthly_avg)
```

```python
# ACF and PACF

lag_acf = acf(monthly_avg,nlags=20)
lag_pacf = pacf(monthly_avg,nlags=20,method='ols')

plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(monthly_avg)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(monthly_avg)),linestyle='--',color='gray')
plt.title('AutoCorrelation Function')

plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(monthly_avg)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(monthly_avg)),linestyle='--',color='gray')
plt.title('Partial AutoCorrelation Function')
```

We can see that PACF and ACF cross first confidence levels at (0,1) interval. Thus we can iteratively choose a model.


```python
# AR Model
model = ARIMA(monthly_avg,order=(1,1,0))
results_AR = model.fit(disp=-1)
plt.plot(monthly_avg)
plt.plot(results_AR.fittedvalues,color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-monthly_avg)**2))
```
